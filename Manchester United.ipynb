#### 주전 선수 비교를 위해 11명 선발
- GK : 1명, CB : 4명, MF : 4명, ST : 2명
- 선발 기준은 (현재 능력치 + 잠재력) / 2
- GK 리스트 : GK
- CB 리스트 : CB, LCB, RCB, LB, RB, LWB, RWB
- MF 리스트 : RCM, LCM, RDM, LDM, CDM, CM, RM, LM, CAM
- ST 리스트 : ST, LW, RW, LF, RF


```python
# 각 포지션 별 리스트 작성

gk_list = ['GK']
cb_list = ['CB', 'LCB', 'RCB', 'LB', 'RB', 'LWB', 'RWB']
mf_list = ['RCM', 'LCM', 'RDM', 'LDM', 'CDM', 'CM', 'RM', 'LM', 'CAM']
st_list = ['ST', 'LW', 'RW', 'LF', 'RF']
```

```python
# for문을 통해 주전 선수 선벌

mu_id = []

# 각 포지션 별 선발 제한 인원 입력
gk_count = 1
cb_count = 4
mf_count = 4
st_count = 2

for i in MU.index:
    if MU['Position'][i] in gk_list:
        if gk_count != 0:
            mu_id.append(MU['Name'][i])
            gk_count -= 1
            MU['Position'][i] = 'GK'
            
    elif MU['Position'][i] in cb_list:
        if cb_count != 0:
            mu_id.append(MU['Name'][i])
            cb_count -= 1
            MU['Position'][i] = 'CB'
            
    elif MU['Position'][i] in mf_list:
        if mf_count != 0:
            mu_id.append(MU['Name'][i])
            mf_count -= 1
            MU['Position'][i] = 'MF'
            
    else:
        if st_count != 0:
            mu_id.append(MU['Name'][i])
            st_count -= 1
            MU['Position'][i] = 'ST'
        
        
  
```

```python
mu_id
```

```python
# 새로운 MU 데이터에 추출된 데이터만 넣기

MU = MU[MU['Name'].isin(mu_id)]
```

## 5-2 영입할 선수 결정

- 솔샤르 감독의 영입 방침에 따라 영입할 선수 결정
- 나이는 어릴수록 좋음
- 잠재력보다는 즉시 전력감 선호
- 포지션은 Mata와 Smalling 선수의 포지션인 RM, CB를 따른다.

```python
market.head()
```

```python
# 영입할 선수 리스트만 있는 데이터 생성

market = data[(data['Position']=='RM') | (data['Position'] == 'CB')]
```

```python
# 이적료 수치 변환

market['Value'] = market['Value'].str.replace('M', '000000')
market['Value'] = market['Value'].str.replace('K', '000')

# '유로 단위' 삭제
market['Value'] = market['Value'].str.slice(1,)
```

```python
# '.' 삭제 후 변환된 단위를 맞추기 위해 나누기 10을 해줌

for i in market.index:
    if '.' in market['Value'][i]:
        market['Value'][i] = int(market['Value'][i].replace('.',''))//10
```

```python
# 전체 데이터를 int로 변환

market['Value'] = market['Value'].astype(int)
```

```python
import matplotlib.pyplot as plt
```

```python
# subplot을 그려 데이터별 선수 확인

f, ax= plt.subplots(2, 4, figsize=(20,10))

vs_list = ['Age', 'Overall', 'Potential' , 'Value']

for i in range(8):
    if i < 4:
        
        # 평균보다 큰 항목에 대해 색깔 바꾸기 
        colors = ['firebrick' if x > market[market['Position']=='CB'][:10][vs_list[i]].mean() else 'gray' for x in market[market['Position']=='CB'][:10][vs_list[i]]]
        
        sns.barplot(data=market[market['Position']=='CB'].iloc[:10], x = vs_list[i], y='Name', ax= ax[i//4, i%4], palette = colors)
        
        # 평균 선 긋기
        ax[i//4, i%4].axvline(market[market['Position']=="CB"][:10][vs_list[i]].mean(), ls = '--')
    
    else:
        sns.barplot(data=market[market['Position']=='RM'].iloc[:10], x = vs_list[i%4], y='Name', ax= ax[i//4, i%4])
        
        ax[i//4, i%4].axvline(market[market['Position']=="RM"][:10][vs_list[i%4]].mean(), ls = '--')
```
